### Distribution Plot Of Numeric columns

```python
numerics_column=numerics.columns
def plotKdeplot(df):
    count=0
    for i in range(int(len(numerics_column)/2)):
        for j in range(2):
            plt.subplot(i+1,j+1,j+1)
            sns.kdeplot(numerics[numerics_column[count]])
            count=count+1;
            plt.show()
```

```python
numerics_column=numerics.columns
def plotBoxplot(df):
    count=0
    for i in range(int(len(numerics_column)/2)):
        for j in range(2):
            plt.subplot(i+1,j+1,j+1)
            sns.boxplot(x=numerics_column[count],data=df)
            count=count+1;
            plt.show()
```

```python
def plot_BoxPlot_AccordingTo_Region(df):
    count=0
    plt.figure(figsize=(12, 6))
    for i in range(int(len(numerics_column)/2)):
        for j in range(2):
            plt.subplot(i+1,j+1,j+1)
            sns.boxplot(x=numerics_column[count],hue="Region",data=df)
            plt.legend(title='Region', fontsize=10, title_fontsize=12, loc='upper left', bbox_to_anchor=(1.05, 1))
            count=count+1;
            plt.show()
    
```

# Handled Outlier Using IQR Method ON the BASIS OF REGION

```python
numerics_column=numerics.columns
def ImputeOutlier_by_Region(x):
    q1,q3=x.quantile([0.25,0.75])
    IQR=q3-q1
    min=q1-1.5*IQR
    max=q3+1.5*IQR
    x=np.where(
        x>max,max,
        np.where(x<min,min,x)
    )
    return x
```

```python
for col in numerics_column:
    df[col]=df.groupby('Region')[col].transform(ImputeOutlier_by_Region)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=404)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train.head(2)
```

```python
Handled_skewed_Transformer=ColumnTransformer(transformers=[  
        ("tf2",PowerTransformer(method='yeo-johnson',standardize=True),['Standard Error','Family',
                                                                        'Freedom','Trust (Government Corruption)',
                                                                        'Generosity'])
    ],remainder="passthrough")
```

```python
Encoded_transformer=ColumnTransformer(transformers=[
    ('tf1',OneHotEncoder(drop="first",sparse_output=False),[5])
],remainder="passthrough")
```

```python
RandomForest_Algo=RandomForestRegressor(bootstrap=True, max_features=0.7,max_samples=0.6,n_estimators=300)
```

```python
BaggingAlgo=BaggingRegressor(bootstrap=True,
 bootstrap_features= False,
 estimator= LinearRegression(),
 max_features= 0.6,
 max_samples= 0.3,
 n_estimators= 100,
oob_score=True)
```

## Applying Bagging Regressor 

```python
pipe1=Pipeline([
    ('tf1',Handled_skewed_Transformer),
    ('tf2',Encoded_transformer),
    ('tf3',BaggingAlgo)   
])
```

```python
pipe1.fit(x_train,y_train)
```

```python
y_pred1=pipe1.predict(x_test)
```

```python
r2_score(y_test,y_pred1)
```
