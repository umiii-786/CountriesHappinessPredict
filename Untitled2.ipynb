```python
df['HappinessGrowth']=df.groupby('Country')['Happiness Score'].transform(findIncreased)
def findIncreased(x):
    data=x.values
    differences = np.diff(data)
    percentage_change = (differences / data[:-1]) * 100
    total = np.sum(percentage_change)
    return total;   
```
