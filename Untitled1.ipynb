```python
def createRegion(df):
    def find_region(country):
        regionsArray = combined_data[combined_data['Country'].str.contains(country, na=False)]['Region'].values
        return regionsArray[0] if len(regionsArray) > 0 else np.nan  # Handle empty arrays

    df['Region'] = df['Country or region'].apply(find_region)
    return df
```

```python
numerics_column=numerics.columns
def plotKdeplot(df):
    count=0
    for i in range(int(len(numerics_column)/2)):
        for j in range(2):
            plt.subplot(i+1,j+1,j+1)
            sns.kdeplot(numerics[numerics_column[count]])
            count=count+1;
            plt.show()
```

```python
numerics_column=numerics.columns
def plotBoxplot(df):
    count=0
    for i in range(int(len(numerics_column)/2)):
        for j in range(2):
            plt.subplot(i+1,j+1,j+1)
            sns.boxplot(numerics[numerics_column[count]])
            count=count+1;
            plt.show()
```

```python
plotBoxplot(numerics)
```

```python
def plot_BoxPlot_AccordingTo_Region(df):
    count=0
    plt.figure(figsize=(12, 6))
    for i in range(int(len(numerics_column)/2)):
        for j in range(2):
            plt.subplot(i+1,j+1,j+1)
            sns.boxplot(x=numerics_column[count],hue="Region",data=df)
            plt.legend(title='Region', fontsize=10, title_fontsize=12, loc='upper left', bbox_to_anchor=(1.05, 1))
            count=count+1;
            plt.show()
```

```python
plot_BoxPlot_AccordingTo_Region(data_18_19)
```

```python
numerics_column=numerics.columns
def ImputeOutlier_by_Region(x):
    q1,q3=x.quantile([0.25,0.75])
    IQR=q3-q1
    min=q1-1.5*IQR
    max=q3+1.5*IQR
    x=np.where(
        x>max,max,
        np.where(x<min,min,x)
    )
    return x
```

```python
for col in numerics_column:
    data_18_19[col]=data_18_19.groupby('Region')[col].transform(ImputeOutlier_by_Region)
```
